## Defanging an IP adress
https://leetcode.com/problems/defanging-an-ip-address/description/

```python
address = input()
print(address.replace(".","[.]"))
```

```python
def address():
    ip_address = input()
    output = ""
    for i in ip_address:
        if i == ".":
            output += "[.]"
        else:
            output += i
    return output
```

```python
address()
```

```python
def even_number_digits(nums):
    count = 0
    for num in nums:
        length = len(str(num))
        if length % 2 == 0:
            count += 1
    return count
```

```python
even_number_digits([12,345,2,6,7896])
```

## Number of Good Pairs
https://leetcode.com/problems/number-of-good-pairs/description/

```python
nums = [1,2,3,1,1,3]
count = 0
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i] == nums[j]:
            count += 1
print(count)
```

## How Many Numbers Are Smaller Than the Current Number
https://leetcode.com/problems/how-many-numbers-are-smaller-than-the-current-number/description/

```python
nums = [8,1,2,2,3]
for i in nums:
    for j in nums:
        print(i,j)
```

```python
def smallerNumbersThanCurrent(nums):
    output = []
    for i in nums:
        count = 0
        for j in nums:
            if i > j:
                count += 1
        output.append(count)
    return output
```

```python
smallerNumbersThanCurrent([8,1,2,2,3])
```

```python
smallerNumbersThanCurrent([6,5,4,8])
```

```python
smallerNumbersThanCurrent([7,7,7,7])
```

## Subtract the Product and Sum of Digits of an Integer
https://leetcode.com/problems/subtract-the-product-and-sum-of-digits-of-an-integer/description/

```python
n = 4421
mul = 1
sum_ = 0
while n > 0:
    rem = n % 10
    mul = mul * rem
    sum_ = sum_ + rem
    n = n // 10
print(mul - sum_)
```

## XOR Operation in an array
https://leetcode.com/problems/xor-operation-in-an-array/description/

```python
n = 4
start = 3
output = []
output.append(start)
for i in range(n-1):
    start += 2
    output.append(start)
print(output)
```

```python
count = output[0]
for j in range(1,4):
    count ^= output[j]
print(count)
```

```python
n = 5
start = 0
nums = [0] * n
for i in range(n):
    nums[i] = start+2*i
    count = nums[0]
    for j in range (1,n):
        count ^= nums[j]
print(nums)
print(count)
```
